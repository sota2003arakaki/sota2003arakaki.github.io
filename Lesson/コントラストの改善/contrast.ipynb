```python
from pickleshare import pickle
import numpy as np
import cv2
import matplotlib.pyplot as plt
def gamma(img, gam):
  img = cv2.imread(img)
  hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
  new_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
  h, w, channel = img.shape
  gamma = gam
  #plt.imshow(hsv_img)
  for y in range(h):
    for x in range(w):
      pixcel = 255*(hsv_img[y,x,2]/255)**(1/gamma)
      if pixcel<0:
        pixcel=0
      elif pixcel > 255:
        pixcel=255
      new_img[y,x,2] = pixcel

  new_img = cv2.cvtColor(new_img, cv2.COLOR_HSV2RGB)
  plt.imshow(new_img)
```

```python
img_l=[]
img_l2=[]
for i in range(1,6):
  img_l2.append("PATH")
  img_l.append("PATH")
```

```python
def gamma2(img, out_img, gan):
  img = cv2.imread(img)
  hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
  new_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
  h, w, channel = img.shape
  gamma = gan
  #plt.imshow(hsv_img)
  lookup_table = np.zeros(256,np.float32)
  for i in range(0,256):
    lookup_table[i]=255.0*pow(float(i)/255.0,1.0/gamma)

  new_img[:,:,2] = cv2.LUT(img[:,:,2],lookup_table)
  new_img = cv2.cvtColor(new_img, cv2.COLOR_HSV2RGB)
  plt.imshow(new_img)
  new_img = cv2.cvtColor(new_img, cv2.COLOR_RGB2BGR)
  cv2.imwrite(out_img,new_img)
```
